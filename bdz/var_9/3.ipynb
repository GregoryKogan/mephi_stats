```python
n = len(X)
sigma = [53, 30, 80]
print(X.std())
for sigma0 in sigma:
    z = X.var() * (n-1) / sigma0**(2)
    p_value = 2*min([sps.chi2.cdf(z, n - 1), 1 - sps.chi2.cdf(z, n - 1)]) 
    print(" chi2-test : The z-statistic is %.3f and the p-value is %.3f." % (z,  p_value))
```
